```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score
from scipy.interpolate import interp1d

# Ustalanie ziarna dla powtarzalności
np.random.seed(0)

# Generowanie sinusoidalnego sygnału z dodatkiem szumu
X_original = np.linspace(0, 10, 100)
y_original = np.sin(X_original)

# Generowanie punktów zróżnicowanych
X_noisy = np.concatenate([
    np.linspace(0, 3, 4),                    # Mniej punktów na początku
    np.linspace(3, 4, 1),                   # Więcej punktów w środku
    np.linspace(4, 7, 2),                   # Trochę punktów nad oryginałem
    np.linspace(7, 10, 3),                   # Trochę punktów pod oryginałem
])
y_noisy = np.sin(X_noisy) + np.random.normal(0, 0.1, len(X_noisy))

# Zmiana kształtu danych dla estymacji wielomianowej
X_noisy = X_noisy[:, np.newaxis]

# Tworzenie modelu wielomianowego o zmiennej złożoności
degrees = [0, 1, 6, 9]  # Różne stopnie wielomianu
colors = ['green', 'red', 'purple', 'blue']

# Wykres oryginalnego sinusoidy z dodatkiem szumu


for degree in degrees:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_noisy, y_noisy)
    y_pred = model.predict(X_original[:, np.newaxis])

    print("Degree:", degree)
    print("Coefficients:", model.named_steps['linearregression'].coef_)
    print("Mean squared error: %.2f" % mean_squared_error(y_original, y_pred))
    print("Coefficient of determination: %.2f" % r2_score(y_original, y_pred))

    # Interpolacja
    f = interp1d(X_original, y_pred, kind='cubic')
    y_smooth = f(X_original)
    plt.scatter(X_noisy, y_noisy, color='black', label="Sygnał spróbkowany")
    plt.plot(X_original, y_original, color='orange', label="Sygnał oryginalny")
    plt.legend()
    plt.plot(X_original, y_smooth, color=colors[degrees.index(degree)], linewidth=2, label="Wielomian %d stopnia" % degree)
    plt.legend()
    plt.xticks(np.arange(0, 11, 1))
    plt.yticks(np.arange(-1, 1.5, 0.5))
    plt.xlim(0, 10)
    plt.ylim(-1.2, 1.2)
    plt.show()


```
